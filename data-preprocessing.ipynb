```python
data['last_updated'] = pd.to_datetime(data['last_updated'])
data['condition_text'] = data['condition_text'].str.strip()
```

```python
data.rename(columns={'temperature_celsius': 'Temperature (°C)', 
                        'temperature_fahrenheit': 'Temperature (°F)', 
                        'wind_kph': 'Wind (kph)', 
                        'humidity': 'Humidity', 
                        'cloud': 'Cloud', 
                        'precip_mm': 'Precipitation (mm)',
                        'wind_mph': 'Wind Speed (mph)',
                        'wind_kph': 'Wind Speed (kph)'}, inplace=True)
```

```python
data.to_csv('cleaned_weather_data.csv')
```
